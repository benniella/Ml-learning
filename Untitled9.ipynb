```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Housing_dataset_train.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['loc'].fillna(df['loc'].mode()[0], inplace = True)
df['title'].fillna(df['title'].mode()[0], inplace = True)
df['bedroom'].fillna(df['bedroom'].mode()[0], inplace = True)
df['bathroom'].fillna(df['bathroom'].mode()[0], inplace = True)
df['parking_space'].fillna(df['parking_space'].mode()[0], inplace = True)
df['price'].fillna(df['price'].mode()[0], inplace =True)

df.isnull().sum()
```

```python
house_type_distribution= df['title'].value_counts()
df
```

```python
house_type_distribution.plot(kind='bar')
plt.title('distribution of house type')
plt.xlabel('house type')
plt.ylabel('count')
plt.show()
```

```python
sns.histplot(df['parking_space'], bins='auto', kde=True)
plt.title('Distribution of Parking Space')
plt.xlabel('Numbers of car Parking Space can Fit')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x=df['parking_space'])
plt.xlabel=('Box Plot for Parking Space')
plt.ylabel={'number of cars parking space can fit'}
plt.show()
```

```python
sns.kdeplot(df['parking_space'], fill = True)
plt.title('kde plot of parking space capacity')
plt.xlabel=('number cars parking space can fit')
plt.show()
```

```python
sns.scatterplot(x='bedroom', y='bathroom', hue='price', data=df, palette='viridis')
plt.title('scatter plot of house price vs bedroom and bathroom')
plt.xlabel('number of bedroom')
plt.ylabel('number of bathroom')
plt.show()
```
